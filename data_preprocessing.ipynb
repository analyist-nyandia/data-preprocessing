```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/blair-developer/dataMain/refs/heads/main/Walmart_Sales.csv')
df.head()
```

**1. Identify missing data and eliminate duplicate entries in the dataset**

```python
#check for missing value
df.isnull().sum()#there is no missing values
```

```python
#eliminate duplicates
#first check for duplicates
df.duplicated().sum()
```

**2. Transform the Date column into a Pandas datetime object.**

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df.head()
```

**3. Add a new column called Revenue, calculated as the product of Weekly_Sales and a hypothetical Fuel_Price column.**

```python
#Add a new column called Revenue
df['Revenue'] = df['Weekly_Sales'] * df['Fuel_Price']
df.head()
```

**4. Identify and handle Garbage values**

```python
df.isnull().sum()#checking for null values
```

```python
df.duplicated().sum()#checking for duplicates
```

**5. Standardize the categorical column `holiday_flag` by mapping the values 0 and 1 to "No Holiday" and "Holiday," respectively.**

```python
# Standardize the 'holiday_flag' column
df['Holiday_Flag'] = df['Holiday_Flag'].map({0: 'No Holiday', 1: 'Holiday'})
df.head()
```
